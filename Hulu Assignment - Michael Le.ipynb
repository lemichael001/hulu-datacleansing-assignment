```python
import pandas as pd
```

```python
#Ran into parser error issue loading the dataframe, so dropped some bad lines

usvideos = pd.read_csv("C:/Users/Michael/Desktop/Hulu Assignment/USvideos.csv", error_bad_lines=False)
uscomments = pd.read_csv("C:/Users/Michael/Desktop/Hulu Assignment/UScomments.csv", error_bad_lines=False)
gbvideos = pd.read_csv("C:/Users/Michael/Desktop/Hulu Assignment/GBvideos.csv", error_bad_lines=False)
gbcomments = pd.read_csv("C:/Users/Michael/Desktop/Hulu Assignment/GBcomments.csv", error_bad_lines=False)
categorylookup = pd.read_csv("C:/Users/Michael/Desktop/Hulu Assignment/categorylookup.csv", error_bad_lines=False)
```

```python
#Add a year to date
usvideos['date'] = usvideos['date'].astype(str) + '.2017'
gbvideos['date'] = usvideos['date'].astype(str) + '.2017'
#Make dates look like actual date format
usvideos['date'] = pd.to_datetime(usvideos['date'])
gbvideos['date'] = pd.to_datetime(usvideos['date'])
```

```python
#Changing quantitative values to numeric
usvideos.astype({'views': int, 'likes': int, 'dislikes': int, 'comment_total':int})
```

```python
#Adding territory column to each dataset
usvideos['territory'] = 'US'
uscomments['territory'] = 'US'
gbvideos['territory'] = 'GB'
gbcomments['territory'] = 'GB'
```

```python
#Join video dataframes with categorylookup so I can have category names instead of numbers
usvideos = pd.merge(usvideos, categorylookup, how = 'left', on = 'category_id')
gbvideos = pd.merge(gbvideos, categorylookup, how = 'left', on = 'category_id')
```

```python
#Stripping '=-' from video_id as it's causing errors
usvideos['video_id'] = usvideos['video_id'].map(lambda x: x.lstrip ('=-'))
gbvideos['video_id'] = gbvideos['video_id'].map(lambda x: x.lstrip ('=-'))
uscomments['video_id'] = uscomments['video_id'].map(lambda x: x.lstrip ('=-'))
gbcomments['video_id'] = gbcomments['video_id'].map(lambda x: x.lstrip ('=-'))
```

```python
#Created a column for prior date to calculate incremental metrics
usvideos['day'] = 1
usvideos['previous_date'] = usvideos['date'] - pd.to_timedelta(usvideos['day'], unit='d')

gbvideos['day'] = 1
gbvideos['previous_date'] = gbvideos['date'] - pd.to_timedelta(gbvideos['day'], unit='d')
```

```python
#Concatenate video_id with date for a unique identifier to get incremental metric
usvideos['video_id_date'] = usvideos['video_id'] + usvideos['date'].map(str)
gbvideos['video_id_date'] = gbvideos['video_id'] + gbvideos['date'].map(str)

usvideos['video_id_previous_date'] = usvideos['video_id'] + usvideos['previous_date'].map(str)
gbvideos['video_id_previous_date'] = gbvideos['video_id'] + gbvideos['previous_date'].map(str)
```

```python
#Creating a 2nd dataframe of video_id_date (primary key) and adding in metrics
usvideosdf = usvideos[['video_id_date', 'views', 'likes', 'dislikes', 'comment_total']]
gbvideosdf = gbvideos[['video_id_date', 'views', 'likes', 'dislikes', 'comment_total']]
```

```python
#Merging 2nd dataframe to original to get prior date metrics
usvideos = pd.merge(usvideos, usvideosdf, how = 'left', left_on = ['video_id_previous_date'], right_on = ['video_id_date'])
gbvideos = pd.merge(gbvideos, gbvideosdf, how = 'left', left_on = ['video_id_previous_date'], right_on = ['video_id_date'])
#Replacing null values with a 0
usvideos.fillna(0, inplace=True)
gbvideos.fillna(0, inplace=True)
```

```python
#Creating fields for daily variance for each metric
usvideos['daily_views'] = usvideos['views_x'] - usvideos['views_y']
usvideos['daily_likes'] = usvideos['likes_x'] - usvideos['likes_y']
usvideos['daily_dislikes'] = usvideos['dislikes_x'] - usvideos['dislikes_y']
usvideos['daily_comment_total'] = usvideos['comment_total_x'] - usvideos['comment_total_y']
gbvideos['daily_views'] = gbvideos['views_x'] - gbvideos['views_y']
gbvideos['daily_likes'] = gbvideos['likes_x'] - gbvideos['likes_y']
gbvideos['daily_dislikes'] = gbvideos['dislikes_x'] - gbvideos['dislikes_y']
gbvideos['daily_comment_total'] = gbvideos['comment_total_x'] - gbvideos['comment_total_y']
```

```python
#Dropping unncessary fields
usvideos = usvideos.drop(['day', 'video_id_previous_date', 'video_id_date_x', 'video_id_date_y', 'views_y', 'likes_y',
              'dislikes_y', 'comment_total_y'], axis=1)
gbvideos = gbvideos.drop(['day', 'video_id_previous_date', 'video_id_date_x', 'video_id_date_y', 'views_y', 'likes_y',
              'dislikes_y', 'comment_total_y'], axis=1)
```

```python
#Union the video and comments dataframes
allvideos = pd.concat([usvideos, gbvideos])
allcomments = pd.concat([uscomments, gbcomments])
```

```python
#Filter to only have values between 9/13 - 9/30
allvideos = allvideos[~(allvideos['date'] < '2017-09-13')]
```

```python
#Export as csv
allvideos.to_csv("allvideos.csv", sep=',')
allcomments.to_csv("allcomments.csv", sep=',')
```
